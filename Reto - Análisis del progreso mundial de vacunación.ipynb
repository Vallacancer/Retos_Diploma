## e. Calcular el promedio de vacunas aplicadas por país.

No me queda muy claro a que se refiere, pero voy a especificar tanto el promedio de vacunas diarias por país, como las vacunas aplicadas por país

```python
import pandas as pd

vacunas_pais = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/country_vaccinations.csv", parse_dates=["date"])

#Vacunas por país
vacunas_por_pais = vacunas_pais["daily_vaccinations"].groupby(vacunas_pais["country"]).sum();
print("Estas son las vacunas aplicadas por país:\n" ,vacunas_por_pais)

#Vacunas diarias por país
vacunas_promedio = vacunas_pais["daily_vaccinations"].groupby(vacunas_pais["country"]).mean();
print("\nEstas son las vacunas promedio por día por país:\n", vacunas_promedio)
```

## g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.

## h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
import pandas as pd

vacunas_pais = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/country_vaccinations.csv", parse_dates=["date"])

#Se busca la fecha más reciente
mayor_dato = vacunas_pais["date"].max();
print("La fecha más reciente es: ", mayor_dato)

#Se busca la fecha más antigüa
menor_dato = vacunas_pais["date"].min();
print("La fecha más antigüa es: ", menor_dato)

#Se calcula la diferencia
delta = mayor_dato - menor_dato;
print("La diferencia entre estas dos es: ", delta)
```

## i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

```python
import pandas as pd

vacunas_pais = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/country_vaccinations.csv", parse_dates=["date"])

conCantidad = vacunas_pais;

#Se concatenan ambas columnas en canVac
conCantidad["canVac"] = conCantidad["vaccines"].str.cat(", " + conCantidad["daily_vaccinations"].astype(str));

#Se muestra como se ve un dato concatenado
print(conCantidad.loc[2,"canVac"])
```

## k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
import pandas as pd

vacunas_pais = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/country_vaccinations.csv", parse_dates=["date"])

pfizer = vacunas_pais[vacunas_pais["vaccines"].str.contains("pfizer", case=False)];

#Nos aseguramos que se haya hecho bien
print(pfizer["vaccines"])
```

## l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.

```python
import pandas as pd

vacunas_pais = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/country_vaccinations.csv", parse_dates=["date"])

#Dataframe conDiferencias
conDiferencias = vacunas_pais.copy();
conDiferencias["diferencias"] = abs(conDiferencias["daily_vaccinations"]-conDiferencias["daily_vaccinations_raw"]);


#Dataframe conCantidad
conCantidad = vacunas_pais.copy();
conCantidad["canVac"] = conCantidad["vaccines"].str.cat(", " + conCantidad["daily_vaccinations"].astype(str));

#Dataframe antes20
antes20 = vacunas_pais.query('date < "2020-12-20"');

#Dataframe pfizer
pfizer = vacunas_pais[vacunas_pais["vaccines"].str.contains("pfizer", case=False)];

with pd.ExcelWriter("/content/drive/MyDrive/Colab Notebooks/DataSets/resultadosReto.xlsx") as writer:
  conDiferencias.to_excel(writer, sheet_name="conDiferencias")
  conCantidad.to_excel(writer, sheet_name="conCantidad")
  antes20.to_excel(writer, sheet_name="antes20")
  pfizer.to_excel(writer, sheet_name="pfizer")
```
