# 7.	Filtra el dataframe para quedarte con 5 países (el más feliz, el menos feliz, México y dos más de tu interés) y visualiza en una misma gráfica los 6 factores. Ejemplo:
image.png

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Se escoge una paleta
sns.set_palette("Set2")

#Se lee el archivo
happiness = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Happiness_report.csv")

#Se crea el dataframe de los países y se coloca el índice
pais_max = happiness["Score"].max()
pais_min = happiness["Score"].min()
paises = happiness.loc[(happiness["Country or region"].isin(["Mexico","United States", "China"]))|
                        (happiness["Score"].isin([pais_max,pais_min]))].set_index("Country or region")

#Se prepara el DataFrame
paises = paises.drop(["Overall rank","Score"],axis=1).T

paises.plot.bar(title = "World Happiness Report");
```

# 8.	Crea una matriz de subgráficas de 2 x 3 con scatter plots del puntaje (score) versus los 6 factores, para determinar qué factor influye más en la evaluación.
image.png

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Se escoge una paleta
sns.set_palette("Set2")

#Se lee el archivo
happiness = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Happiness_report.csv")

#Se crea la figura y las subgráficas
fig, axs = plt.subplots(2,3,figsize=(14,10))
fig.suptitle("Contribution in happiness score")
plt.subplots_adjust(wspace=0.4,hspace=0.4)

#Se obtiene la forma de la subgráfica
filas,columnas = axs.shape
llave = 3

#Se grafica para cada columna
for i in range(filas):
  for j in range(columnas):
    ax_now = axs[i][j]
    sns.scatterplot(x=happiness["Score"],y=happiness.iloc[:,llave],ax=ax_now)
    llave=llave+1
```

# 9.	Combina con el dataframe metada (Metadata.csv) para graficar la felicidad promedio por región.
image.png

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Se escoge una paleta
sns.set_palette("Set2")

#Se lee el archivo
happiness = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Happiness_report.csv")
metada = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Metadata.csv")

#Se combinan los DataFrames
happiness_meta = pd.merge(happiness, metada, left_on="Country or region", right_on="TableName").set_index("Overall rank")

region_felicidad = happiness_meta[["Score", "Region"]].groupby("Region").mean()

#Se grafica
sns.barplot(x=region_felicidad.index, y="Score", data=region_felicidad)
plt.xticks(rotation=90);
plt.title("World Happiness Report");
```
