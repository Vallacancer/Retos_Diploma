## 4.	Conjunta de manera correcta la información, de tal forma que queden todos los registros en un solo DataFrame que contenga todas las columnas necesarias.

## 6.	Ejecuta la instrucción adecuada en Pandas para encontrar lo siguiente:

### a.	El porcentaje de personas que usan Windows, Mac o Linux, no registrado como su sistema operativo en su computadora

```python
import pandas as pd

#Se llen y guardan los DF
empleados1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Test.csv")
empleados1_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Test2.csv")
empleados2 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Train.csv")
empleados2_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Train2.csv")

#Se conjuntan los empleados, cuidando que los índices se ignoren
empleados_total = pd.concat([empleados1, empleados2], ignore_index = True, copy=False);
#Se hace lo mismo para la información de los equipos de los empleados
equipos_total = pd.concat([empleados1_data, empleados2_data], ignore_index = True, copy = False);

#Se hace el DataFrame final, conjuntando a los empleados con las especificaciones de sus equipos
data_total = pd.merge(empleados_total, equipos_total, on="Employee_ID", how="left", copy=False);

#Se buscan los valores nulos y se actualizan
data_total.loc[data_total[" Computer_OS"].isnull()," Computer_OS"]="no registrado"
data_total.loc[data_total[" Mobile_OS"].isnull()," Mobile_OS"]="no registrado"

conteo_compuOS = data_total.groupby(" Computer_OS").agg({" Computer_OS":"size"})

#Se renombra
conteo_compuOS = conteo_compuOS.rename(columns={" Computer_OS": "Conteo"})

#Se reinicia el índice
conteo_compuOS = conteo_compuOS.reset_index()

conteo_compuOS["Porcentaje de Usuarios"] = conteo_compuOS["Conteo"]*100/ conteo_compuOS["Conteo"].sum()
conteo_compuOS['Porcentaje de Usuarios'] = conteo_compuOS["Porcentaje de Usuarios"].astype(str) + "%"
conteo_compuOS = conteo_compuOS[[' Computer_OS',"Porcentaje de Usuarios"]]

conteo_compuOS
```

### e.	El número total de tickets generados por la gente, con cada sistema operativo de escritorio.

```python
import pandas as pd

#Se llen y guardan los DF
empleados1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Test.csv")
empleados1_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Test2.csv")
empleados2 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Train.csv")
empleados2_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Train2.csv")

#Se conjuntan los empleados, cuidando que los índices se ignoren
empleados_total = pd.concat([empleados1, empleados2], ignore_index = True, copy=False);
#Se hace lo mismo para la información de los equipos de los empleados
equipos_total = pd.concat([empleados1_data, empleados2_data], ignore_index = True, copy = False);

#Se hace el DataFrame final, conjuntando a los empleados con las especificaciones de sus equipos
data_total = pd.merge(empleados_total, equipos_total, on="Employee_ID", how="left", copy=False);

#Se buscan los valores nulos y se actualizan
data_total.loc[data_total[" Computer_OS"].isnull()," Computer_OS"]="no registrado"
data_total.loc[data_total[" Mobile_OS"].isnull()," Mobile_OS"]="no registrado"

#Se agrupan los valores de los tickets por OS de las compus
tickets_totales = data_total[[" Computer_tickets"," Mobile_tickets"]].groupby(data_total[" Computer_OS"]).sum()

#Se suman los valores de los tickets y se guarda únicamente el total
tickets_totales["Tickets_totales"] = tickets_totales[' Computer_tickets'] + tickets_totales[" Mobile_tickets"]
tickets_totales = pd.DataFrame(tickets_totales["Tickets_totales"])

tickets_totales
```

## 7.	Envía el resultado de cada solicitud de información a un archivo .xlsx, donde el resultado de cada consulta esté en su propia hoja de trabajo.

```python
import pandas as pd

#Se llen y guardan los DF
empleados1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Test.csv")
empleados1_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Test2.csv")
empleados2 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Train.csv")
empleados2_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/Train2.csv")

#Se conjuntan los empleados, cuidando que los índices se ignoren
empleados_total = pd.concat([empleados1, empleados2], ignore_index = True, copy=False);
#Se hace lo mismo para la información de los equipos de los empleados
equipos_total = pd.concat([empleados1_data, empleados2_data], ignore_index = True, copy = False);

#Se hace el DataFrame final, conjuntando a los empleados con las especificaciones de sus equipos
data_total = pd.merge(empleados_total, equipos_total, on="Employee_ID", how="left", copy=False);

#Se buscan los valores nulos y se actualizan
data_total.loc[data_total[" Computer_OS"].isnull()," Computer_OS"]="no registrado"
data_total.loc[data_total[" Mobile_OS"].isnull()," Mobile_OS"]="no registrado"

#Solicitud a
conteo_compuOS = data_total.groupby(" Computer_OS").agg({" Computer_OS":"size"})
conteo_compuOS = conteo_compuOS.rename(columns={" Computer_OS": "Conteo"})
conteo_compuOS = conteo_compuOS.reset_index()
conteo_compuOS["Porcentaje de Usuarios"] = conteo_compuOS["Conteo"]*100/ conteo_compuOS["Conteo"].sum()
conteo_compuOS['Porcentaje de Usuarios'] = conteo_compuOS["Porcentaje de Usuarios"].astype(str) + "%"
conteo_compuOS = conteo_compuOS[[' Computer_OS',"Porcentaje de Usuarios"]]

#Solicitud b
conteo_celOS = data_total.groupby(" Mobile_OS").agg({" Mobile_OS":"size"})
conteo_celOS = conteo_celOS.rename(columns={" Mobile_OS": "Conteo"})
conteo_celOS = conteo_celOS.reset_index()
conteo_celOS["Porcentaje de Usuarios"] = conteo_celOS["Conteo"]*100/ conteo_celOS["Conteo"].sum()
conteo_celOS['Porcentaje de Usuarios'] = conteo_celOS["Porcentaje de Usuarios"].astype(str) + "%"
conteo_celOS = conteo_celOS[[' Mobile_OS',"Porcentaje de Usuarios"]]

#Solicitud c
edad_promedio = pd.DataFrame(data_total["Age"].groupby(data_total[" Computer_OS"]).mean())

#Solicitud d
educacion_promedio = pd.DataFrame(data_total["Education_Level"].groupby(data_total[" Computer_OS"]).mean())

#Solicitud e
tickets_totales = data_total[[" Computer_tickets"," Mobile_tickets"]].groupby(data_total[" Computer_OS"]).sum()
tickets_totales["Tickets_totales"] = tickets_totales[' Computer_tickets'] + tickets_totales[" Mobile_tickets"]
tickets_totales = pd.DataFrame(tickets_totales["Tickets_totales"])

with pd.ExcelWriter("/content/drive/MyDrive/Colab Notebooks/DataSets/DS_C2_SC3_SantiagoValladares.xlsx") as writer:
  data_total.to_excel(writer,sheet_name="Base de datos")
  conteo_compuOS.to_excel(writer,sheet_name="Solicitud a")
  conteo_celOS.to_excel(writer,sheet_name="Solicitud b")
  edad_promedio.to_excel(writer,sheet_name="Solicitud c")
  educacion_promedio.to_excel(writer,sheet_name="Solicitud d")
  tickets_totales.to_excel(writer,sheet_name="Solicitud e")
```
