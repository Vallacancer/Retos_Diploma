## Paso 2, Sección 1 (menor a 9480 dólares)

```python
import pandas as pd
import math

df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DataSets/Insurance-Reto.xlsx", sheet_name="Datos Costo Póliza")

df
```

```python
poliza = df["Costo Póliza"]

tn1 = (poliza.mean() - 9480)/(poliza.std()/math.sqrt(len(poliza)))
```

```python
from scipy.stats import t

t.cdf(tn1, (len(poliza)-1))
```

## Sección 2 Parte 2 (Mujeres vs hombres)

```python
mujeres = df.loc[df["Género"]=="female", "Indice de masa corporal"]
hombres = df.loc[df["Género"]=="male", "Indice de masa corporal"]

nominador = (mujeres.mean()-hombres.mean())
denominador = math.sqrt((mujeres.std()**2/len(mujeres))+(hombres.std()**2/len(hombres)))

tn = nominador/denominador

t.cdf(tn,(len(mujeres)+len(hombres)-2))
```

## Sección 2 Parte 3 (Hijos y costo)

```python
df["Cantidad de hijos"].unique()
```

```python
from scipy.stats import f_oneway

hijo0 = df.loc[df["Cantidad de hijos"]==0,"Costo Póliza"]
hijo1 = df.loc[df["Cantidad de hijos"]==1,"Costo Póliza"]
hijo2 = df.loc[df["Cantidad de hijos"]==2,"Costo Póliza"]

estadistico_f, valor_p = f_oneway(hijo0,hijo1,hijo2)

print("El valor estadístico de la prueba Fisher es: ", estadistico_f)
print("El valor-p de la prueba es: ", valor_p)
```

## Sección 2, Parte 4 (Hijos y género)

```python
hijos_hombre = df.loc[df["Género"]=="male","Cantidad de hijos"].value_counts()
hijos_mujer = df.loc[df["Género"]=="female","Cantidad de hijos"].value_counts()

tabla_chi2 = pd.DataFrame({"0":[hijos_hombre[0],hijos_mujer[0]],
                           "1":[hijos_hombre[1],hijos_mujer[1]],
                           "2":[hijos_hombre[2],hijos_mujer[2]]}, index = ("hombre","mujer"))

tabla_chi2
```

```python
from scipy.stats import chi2_contingency

stat, valor_p, grados, esperado = chi2_contingency(tabla_chi2)

esperado
```

```python
print("El valor estadístico de la prueba es: ", stat)
print("El valor-p de la prueba es: ", valor_p)
print("Los grados de libertad son: ", grados)
```

```python
fumador = df.loc[df["Fumador"]=="yes", "Costo Póliza"]
nofumador = df.loc[df["Fumador"]=="no", "Costo Póliza"]

nominador = (fumador.mean()-nofumador.mean())
denominador = math.sqrt((fumador.std()**2/len(fumador))+(nofumador.std()**2/len(nofumador)))

tn = nominador/denominador

glnominador = math.pow((fumador.std()**2/len(fumador))+(nofumador.std()**2/len(nofumador)),2)
gldenominador = (((1/(len(fumador)-1))*math.pow((fumador.std()**2/len(fumador)),2))
+((1/(len(nofumador)-1))*math.pow(nofumador.std()**2/len(nofumador),2)))
gl = glnominador / gldenominador

valor_p = 2*t.sf(tn,gl)
```

```python
print("El valor-p de la prueba es: ", valor_p)
```
